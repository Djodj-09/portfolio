### 2) Loading data

```python
def load(path):
    df = pd.read_csv(path)
    return df
```

```python
df_c = load("/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_customers_dataset.csv")
df_g = load("/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_geolocation_dataset.csv")
df_it = load("/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_order_items_dataset.csv")
df_or_pay = load("/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_order_payments_dataset.csv")
df_or_rev = load("/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_order_reviews_dataset.csv")
df_or_ds = load("/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_orders_dataset.csv")
df_p = load("/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_products_dataset.csv")
df_s = load("/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/olist_sellers_dataset.csv")
df_cat = load("/kaggle/input/datasets/organizations/olistbr/brazilian-ecommerce/product_category_name_translation.csv")

print("Loading data")
print(f"customer : {df_c.shape}")
print(f"Geolocalisation : {df_g.shape}")
print(f"order item : {df_it.shape}")
print(f"order payment : {df_or_pay.shape}")
print(f"order review : {df_or_rev.shape}")
print(f"order dataset : {df_or_ds.shape}")
print(f"products : {df_p.shape}")
print(f"sellers : {df_s.shape}")
print(f"product categoy : {df_cat.shape}")
```
