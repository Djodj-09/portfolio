# Webscraping : Share price Netflix ans Amazon

# Objectives
- use yahoo finance website to  Extract Historical Share Price of Netflix and Amazon's corporation .

### import module

```python
import pandas as pd
import yfinance as yf
important yfinance as yf
```

# 1) For Netflix

### Extract data 
For the company Netflix the Ticker symbol is NFLX

```python
nflx  = yf.Ticker("NFLX")
```

### Extracting Share Price
- A share is the single smallest part of a company's stock  that you can buy, the prices of these shares fluctuate over time.

```python
nflx_data = nflx.history(period="max")
```

```python
nflx_data.tail()
```

```python
nflx_data.reset_index(inplace=True)
nflx_data.plot(x= 'Date', y='Open')
```

```python
nflx_data.Dividends
```

# 2) For Amazon

### Extract data 
For the company Amazon,  the Ticker symb is AMZN

```python
amzn  = yf.Ticker("AMZN")
```

### Extracting Share Price
